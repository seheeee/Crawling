# 1. 동적 vs. 정적페이지
- 동적 페이지
    - 데이터의 추가적인 변경이 일어나는 페이지
    - 응답받은 HTML에 원하는 정보가 없음
    - <u>자바스크립트</u>가 내용을 가지고 있음 > <u>웹브라우저를 열어야하기 때문에 selenium 사용</u>
- 정적 페이지
    - 데이터의 추가적인 변경이 일어나지 않는 페이지
    - 응답받은 HTML에 원하는 정보가 들어있음
- 구별 방법
    - 개발자도구(f12) > 설정 > Debugger (check)Disable JavaScript > 새로고침 
    <br> -> JavaScript가 기능 껐을 때의 정적 페이지를 확인 가능

### - 동적 크롤링

```python
from selenium import webdriver
from bs4 import BeautifulSoup

driver = webdriver.Chrome()
driver.get('https://startcoding.pythonanywhere.com/dynamic')

#html 추출
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser') 
title = soup.select_one('.product-name').text
print(title)

driver.close()
```

# 3. 셀레니움을 사용한 스크롤 내리기

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

driver = webdriver.Chrome()

driver.get('https://startcoding.pythonanywhere.com/dynamic')
```

```python
# 절대 위치로 스크롤
driver.execute_script("window.scrollTo(0,1000)") #자바스크립트 실행 명령어 #x좌표 0, y좌표 1000만큼 스크롤
```

```python
# 상대 위치로 스크롤 #현재 위치 기준으로 스크롤
driver.execute_script("window.scrollBy(0,1000)")
```

```python
# 현재 스크롤 높이 #전체 바디태그의 높이
driver.execute_script("return document.body.scrollHeight")
```

```python
# 페이지끝까지 스크롤
driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
```

# 4. 심화

## 1) 특정요소로 스크롤

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

driver = webdriver.Chrome()

driver.get('https://startcoding.pythonanywhere.com/dynamic')

tag = driver.find_element(By.CSS_SELECTOR, '#aside > div:nth-child(3) > div > div:nth-child(6) > label')
driver.execute_script("arguments[0].scrollIntoView(true)",tag) #[0] 자리에 tag가 들어가는 구조
```

## 2) 특정요소가 나타날때까지 스크롤

```python
import time

while True:
    try:
        element = driver.find_element(By.CSS_SELECTOR, '#product-container > div:nth-child(22) > div > div.product-body > h3 > a')
        print(element.text)
        break
    except:
        driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
        time.sleep(1)
```

```python
last_height = driver.execute_script("return document.body.scrollHeight")

# 왜 한번만 스크롤 될까?
# while True:
#     driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
#     cur_height = driver.execute_script("return document.body.scrollHeight")
#     if cur_height == last_height:
#         break
#     last_height = cur_height

import time
# 왜 한번만 스크롤 될까? > 스크롤하고 로딩 시간 필요
while True:
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    time.sleep(1)

    cur_height = driver.execute_script("return document.body.scrollHeight")
    if cur_height == last_height:
        break
    last_height = cur_height
    
```
